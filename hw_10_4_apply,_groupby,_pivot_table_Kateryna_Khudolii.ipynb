1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
branches_income = pd.pivot_table(data=df,values='gross income',aggfunc='sum',index='Branch')
branches_income['gross income'] = round(branches_income['gross income'],2)
branches_income = branches_income.sort_values(by='gross income',ascending=False)
branches_income
```

Найбільш прибутковим є філіал С.

2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  
Знайдіть відповідь за допомогою функціоналу Pandas.

```python
branches_by_cities = pd.pivot_table(data=df,values='gross income',aggfunc='sum',index='Branch',columns='City')
branches_by_cities
```

1. Питання 1 - Який середній рейтинг покупок у різних філіалах?

```python
avg_rating = pd.pivot_table(data=df,index='Branch',values='Rating',aggfunc='mean')
avg_rating
```

Бачимо, що найвищий рейтинг у найприбутковішому філіалі С, а найнижчий - в найменш прибутковому філіалі В. Тобто, філіалу В варто попрацювати над якістю обслуговування клієнтів, щоб підвищити свій прибуток.

2. Питання 2 - Які категорії клієнтів частіше роблять покупки в різних філіалах?

```python
categ_per_branch = pd.pivot_table(data=df,index='Branch',columns='Customer type',values='Invoice ID',aggfunc='count')
categ_per_branch
```

Бачимо, що кількість покупок від членів програми лояльності та звичайних клієнтів майже однакова, отже цей фактор не впливає значним чином на прибуток.

3. Питання 3 - який середній чек в кожному з філіалів?

```python
average_check = pd.pivot_table(data=df,index='Branch',values='Total',aggfunc='mean')
average_check
```

Бачимо, що середній чек в найприбутковішому філіалі С найвищий. Тобто, для підвищення прибутку іншим філіалам варто попрацювати над збільшенням середнього чека (розширення продуктової лінійки, акції на кшталт 1+1=3, тощо).
