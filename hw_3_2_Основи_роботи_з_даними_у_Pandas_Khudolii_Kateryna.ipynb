7.2 Сформулюйте 3 гіпотези, чому в одному місті gross income вище, як можна дізнатись чи правдива ця гіпотеза з використанням даних (наявних і додаткових), та як висновки з цих гіпотез можна використати для прийняття рішень як підвищити прибуток в мережі супермаркетів. Відповідь напишіть текстом в окремій клітинці типу Markdown.

Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
print("Average gross income in Yangon:")
print(df[df.City=="Yangon"]["gross income"].mean())
print("Average gross income in Mandalay:")
print(df[df.City=="Mandalay"]["gross income"].mean())
```

Чому середній валовий дохід в місті Mandalay вищий за середній валовий дохід в місті Yangon?

Гіпотеза 1: в місті Yangon продається менше товарів більш "прибуткових" категорій, ніж в місті Mandalay, або середній чек в цих категоріях менший. Перевірити це можна, перевіривши середній валовий дохід за категоріями:

```python
df.groupby(['Product line','City'])['gross income'].mean()
```

Дійсно бачимо, що в Mandalay на порядок вищий середній валовий дохід в категорії Health and beauty. Розглянемо кількість інвойсів в категоріях:

```python
df.groupby(['Product line','City'])['Invoice ID'].count()
```

Як бачимо, річ дійсно в розмірі середнього чека в певній категорії - кількість продажів по категоріях приблизно однакова. Що ж може впливати на середній чек? Гіпотеза #2 - в місті Mandalay більше покупок здійснюють members, аніж normal customers.

```python
df.groupby(['Product line','City', 'Customer type'])['Invoice ID'].count()
```

Ця гіпотеза теж не підтвердилась - кількість покупок між member та normal приблизно однакова.
Гіпотеза #3 - в місті Mandalay середня ціна товарів, що продаються, більша.

```python
df.groupby(['Product line','City'])['Unit price'].mean()
```

І дійсно це так. В найбільш "дорогій" категорії в місті Mandalay товари в середньому дорожчі на 9 валютних одиниць. Чи це повʼязано з більш складною логістикою, чи більш преміальним сегментом товарів, представленим в магазині, можна дізнатись із додаткових даних.

Одна із гіпотез підтвердилась, в подальшому я б ще перевірила розмір податку (є країни, де податок залежить від області або штату), рейтинг (можливо більш задоволені покупці частіше повертаються і більше купують, і можливо в Mandalay краща якість обслуговування), середній валовий дохід по різним типам покупців (можливо, в Mandalay частіше купують жінки, тому товарів з beauty категорії продається більше).
